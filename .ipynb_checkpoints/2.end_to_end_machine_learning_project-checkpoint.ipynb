## 1. 큰 그림 보기

*머신러닝 부동산 회사에 오신 것을 환영한다! 여러분이 할 작업은 캘리포니아 지역 주택의 여러 특성을 사용해 중간 가격을 예측하는 것이다.*

* 블록 그룹은 미국 인구조사국에서 샘플 데이터를 발표하는 데 사용하는 최소한의 지리적 단위이다.
    <br>(하나의 블록그룹은 보통 600 ~ 3,000명의 인구를 나타낸다. 여기서는 간단하게 구역이라고 부른다.)
* 이 데이터로 모델을 학습시켜 다른 측정 데이터가 주어졌을 때 구역의 중간 주택 가격을 예측해야 한다.

**문제 정의**

* 첫번째로 할 질문은 "비지니스의 목적이 정확히 무엇인가?" 이다.
    <br>(아마도 모델 만들기가 최종 목적은 아닐 것이다.)
    <br>답변자는 이 모델의 출력(구역의 중간 주택 가격에 대한 예측)이 여러 가지 다른 신호와 함께 다른 머신러닝 시스템에 입력으로 사용된다고 한다. 
    <br>즉, 뒤따르는 시스템이 해당 지역에 투자할 가치가 있는지 결정한다는 것이다. 
    <br>이 결정이 수익에 직결되기 때문에 올바르게 예측하는 것은 매우 중요하다.


* 두번째 질문은 "현재 솔루션은 어떻게 구성되어 있는가?" 이다.(만약 있다면)
    <br>현재 상황은 문제 해결 방법에 대한 정보는 물론이고 참고 성능으로도 사용할 수 있다.
    <br>답변자는 현재 구역 주택가격을 전문가가 수동으로 추정한다고 알려주었다.
    <br>(이는 비용과 시간이 많이 들고 추정 결과도 썩 좋지 않다. 여러 이유로 인해 회사는 구역의 데이터를 기반으로 중간 주택 가격을 예측하는 모델을 훈련시키는 쪽이 유용하다고 생각한다.)
    

* 시스템 설계 준비
    <br>레이블된 훈련 샘플이 있으니 전형적인 지도 학습 작업이다.
    <br>값을 예측해야 하므로 전형적인 회귀 문제이다. 
    <br>좀 더 구체적으로는 예측에 사용할 특성이 여러 개(구역의 인구, 중간 소득 등)이므로 다중 회귀 문제이다.
    <br>각 구역마다 하나의 값을 예측하므로 단변량 회귀(univariate regression) 문제이다.
    <br>이 시스템으로 들어오는 데이터에 연속적인 흐름이 없으므로 빠르게 변하는 데이터에 적응하지 않아도 되고, 데이터가 메모리에 들어갈 만큼 충분히 작으므로 일반적인 배치 학습이 적절하다.

**성능 측정 지표 선택**

* 회귀 문제의 전형적인 성능 지표는 평균 제곱근 오차(root mean square error)이다.
<br>오차가 커질수록 이 값은 더욱 커지므로 예측에 얼마나 많은 오류가 있는지 가늠하게 해준다.


* 이상치로 보이는 구역이 많을 경우 평균 절대 오차(mean absolute error, 또는 평균 절대 편차(mean absolute deviation)라고도 한다.)를 고려할 수도 있다.


* RMSE와 MAE 모두 예측값의 벡터와 타겟값의 벡터 사이의 거리를 재는 방법이다.
<br>거리 측정에는 여러 가지 방법(또는 norm)이 가능하다.


* norm의 지수가 클수록 큰 값의 원소에 치우치며 작은 값은 무시된다. 그래서 RMSE가 MAE보다 조금 더 이상치에 민감하다.
<br>하지만 (종 모양 분포의 양 끝단처럼) 이상치가 매우 드물면 RMSE가 잘 맞아 일반적으로 널리 사용된다.

```python
# 테스트셋 만들기
# 예시로 만든 것이다. 실전에서는 사이킷런의 train_test_split()를 사용한다.
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
np.random.seed(42)
train_set, test_set = split_train_test(housing, 0.2)
print(len(train_set), len(test_set))
```

프로그램을 다시 실행하면 다른 테스트셋이 생성되므로 처음 실행에서 테스트셋을 저장하고 다음번 실행에서 다시 불러오던가 난수의 초기값을 지정하는 것이다.
<br>하지만 이 두 가지 방법 모두 다음번에 업데이트된 데이터셋을 사용하려고 했을 때 문제가 된다.
<br>데이터셋을 업데이트한 후에도 안정적인 훈련/테스트 분할을 위한 일반적인 해결책은 샘플의 식별자를 사용하여 테스트셋으로 보낼지 말지 정하는 것이다.
<br>(샘플이 고유하고 변경 불가능한 식별자를 가지고 있다고 가정한다.)
<br>예를 들어 각 샘플마다 식별자의 해시값을 계산하여 해시 최대값의 20%보다 작거나 같은 샘플만 테스트셋으로 보낼 수 있다.
<br>이렇게 하면 여러 번 반복 실행되면서 데이터셋이 갱신되더라도 테스트셋이 동일하게 유지된다.
<br>새로운 테스트셋은 새 샘플의 20%를 갖게 되지만 이전에 훈련셋에 있는 샘플은 포함시키지 않을 것이다.

```python
from zlib import crc32
import hashlib

# 파이썬2와 호환성을 위해 비트연산을 하였다.
#def test_set_check(identifier, test_ratio):
#    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32

def test_set_check(identifier, test_ratio, hash=hashlib.md5):
    #print(np.int64(identifier)) 
    #print(hash(np.int64(identifier)))
    #print(hash(np.int64(identifier)).digest())
    #print(bytearray(hash(np.int64(identifier)).digest()))
    #print(bytearray(hash(np.int64(identifier)).digest())[-1]) # 0~255까지의 값을 리턴
    return bytearray(hash(np.int64(identifier)).digest())[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
# 주택 데이터셋에는 식별자 컬럼이 없지만 대신 행의 인덱스를 ID로 사용하면 간단히 해결된다.
housing_with_id = housing.reset_index()   # `index` 열이 추가된 데이터프레임을 반환한다
housing_with_id
```

```python
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

전문가가 중간 소득이 중간 주택 가격을 예측하는 데 매우 중요하다고 이야기 해주었다고 가정하자.
<br>이 경우 테스트 셋이 전체 데이터셋에 있는 여러 소득 카테고리를 잘 대표해야 한다.
<br>중간 소득이 연속적인 숫자형 특성이므로 소득에 대한 카테고리 특성을 만들어야 한다.
<br>중간 소득의 히스토그램을 조금 더 자세히 살펴보자.(위의 multiple plot 확인)
<br>중간 소득 대부분은 1.5에서 6(즉 $15,000 ~ $60,000) 사이에 모여 있지만 일부는 $60,000를 넘기도 한다.
<br>계층별로 데이터셋에 충분한 샘플 수가 있어야 한다. 그렇지 않으면 계층의 중요도를 추정하는데 편향이 발생할 것이다.
<br>이 말은 너무 많은 계층으로 나누면 안된다는 뜻이고 각 계층이 충분히 커야 한다는 것이다.



```python
# 카테고리 5개를 가진 소득 카테고리 특성을 만든다.(1에서 5까지 레이블을 가진다)
# 카테고리 1은 0에서 1.5까지 범위(즉 $15,000 이하)이고 2는 1.5에서 3까지 범위가 되는 식이다.
housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
```

```python
housing["income_cat"].value_counts()
```

```python
housing["income_cat"].hist()
```

```python
# 이제 소득 카테고리를 기반으로 계층 샘플링을 할 준비가 되었으니 사이킷런의 StratifiedShuffleSplit를 사용한다.
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
# 계층 샘플링으로 만든 테스트셋에서 소득 카테고리 비율
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
# 전체 데이터셋에 있는 소득 카테고리의 비율
# 계층 샘플링으로 만든 테스트셋이 전체 데이터셋에 있는 소득 카테고리의 비율과 거의 같다.
housing["income_cat"].value_counts() / len(housing)
```

```python
# 일반 무작위 샘플링으로 만든 테스트셋과 비교해 보면 많이 다른것을 알 수 있다.

def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()

compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100

compare_props
# 전체, 계층 샘플링, 무작위 샘플링, 무작위 샘플링 오류율, 계층 샘플링 오류율
```

```python
# income_cat 특성을 삭제해서 데이터를 원래 상태로 돌려놓는다.
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

**데이터 정제**

앞서 total_bedrooms 특성에 값이 없는 경우를 보았는데 이를 고쳐보겠다.  방법은 세 가지이다.

* 해당 구역을 제거한다.
* 전체 특성을 삭제한다.
* 어떤 값으로 채운다(0, 평균, 중간값 등)

```python
sample_incomplete_rows = housing[housing.isnull().any(axis=1)].head()
sample_incomplete_rows
```

```python
median = housing["total_bedrooms"].median()
sample_incomplete_rows["total_bedrooms"].fillna(median, inplace=True) # 옵션 3
```

옵션 3을 선택하면 훈련셋에서 중간 값을 계산하고 누락된 값을 이 값으로 채워 넣어야 한다.
<br>계산한 중간값을 저장하는 것을 잊지 말라.
<br>나중에 시스템을 평가할 때 테스트셋에 있는 누락된 값과 시스템이 실제 운영될 때 새로운 데이터 있는 누락된 값을 채워넣는데 필요하기 때문이다.
<br>(훈련셋에서 학습한 것이 소용없어지므로 테스트셋의 중간값을 사용하면 안된다.)

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")

# 중간값이 수치형 특성에서만 계산될 수 있기 때문에 텍스트 특성인 ocean_proximity를 제외한 데이터 복사본을 생성한다.
housing_num = housing.drop("ocean_proximity", axis=1)
# 다른 방법: housing_num = housing.select_dtypes(include=[np.number])

# imputer객체의 fit()를 이용해 훈련 데이터에 적용
imputer.fit(housing_num)
```

```python
# imputer는 각 특성의 중간값을 계산해서 그 결과를 객체의 statistics_ 속성에 저장한다.
# total_bedrooms 특성에만 누락된 값이 있지만 나중에 시스템이 서비스될 때 
# 새로운 데이터에서 어떤 값이 누락될지 확인할 수 없으므로 모든 수치형 특성에 imputer를 적용하는 것이 바람직
imputer.statistics_
```

```python
# 각 특성의 중간값이 수동으로 계산한 것과 같은지 확인해 보라.
housing_num.median().values
```

```python
# 이제 학습된 imputer객체를 사용해 훈련셋에서 누락된 값을 학습한 중간값으로 바꿀 수 있다.
X = imputer.transform(housing_num)
```

```python
# 결과는 변환된 특성들이 들어있는 평범한 넘파이 배열이다. 이를 다시 pandas의 dataframe으로 되돌릴 수 있다.
housing_tr = pd.DataFrame(X, columns=housing_num.columns,index=housing_num.index)

housing_tr.loc[sample_incomplete_rows.index.values]
```

**나만의 변환기**

특별한 정제 작업이나 어떤 특성들을 조합하는 등의 작업을 위해 자신만의 변환기를 만들어야 할 때가 있다.
<br>내가 만든 변환기를 (파이프라인과 같은)사이킷런의 기능과 매끄럽게 연동하고 싶을 것이다.
<br>사이킷런은 (상속이 아닌) 덕 타이핑(duck typing)을 지원하므로 fit() (self를 반환), transform(), fit_transform()을 구현한 파이썬 클래스를 만들면 된다.
<br>(덕 타이핑은 상속이나 인터페이스 구현이 아니라 객체의 속성이나 메서드가 객체의 유형을 결정하는 방식을 말한다.)


마지막 메서드는 TransformMixin을 상속하면 자동으로 생성된다.
<br>또한 BaseEstimator를 상속하면 (그리고 생성자에 `*args`나 `**kargs`를 사용하지 않으면) 하이퍼 파라미터 튜닝에 필요한 두 메서드(get_params()와 set_params())를 추가로 얻게 된다.


<br>(파이썬에서 이름에 Mixin이 있으면 객체의 기능을 확장하려는 목적으로 만들어진 클래스를 나타낸다.
<br>TransformerMixin은 fit_transform() 하나를 가지고 있으며 이를 상속하는 모든 파이썬 클래스에 이 메서드를 제공한다.
<br>TransformerMixin의 fit_transform()은 단순히 fit()과 transform()을 연결(method chaining)한 것이다.)


<br>(get_params()와 set_params()는 사이킷런의 파이프라인과 그리드 탐색에 꼭 필요한 메서드이므로 모든 추정기와 변환기는
<br>BaseEstimator를 상속해야 한다. 이 두 메서드는 생성자에 명시된 매개변수만을 참조하므로
<br>`*args` 나 `**kargs`는 사용할 수 없다.

```python
from sklearn.base import BaseEstimator, TransformerMixin

# 열 인덱스
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # *args 또는 **kargs 없음
        self.add_bedrooms_per_room = add_bedrooms_per_room
        
    def fit(self, X, y=None):
        return self  # 아무것도 하지 않는다
    
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.to_numpy())
```

이 경우는 변환기가 add_bedrooms_per_room 하이퍼 파라미터 하나를 가지며 기본값을 True로 지정한다.
<br>(합리적인 기본값을 주는 것이 좋다.)
<br>이 특성을 추가하는 것이 머신러닝 알고리즘에 도움이 될지 안될지 이 하이퍼파라미터로 쉽게 확인해 볼 수 있다.
<br>일반적으로 100% 확신이 없는 모든 데이터 준비 단계에 대해 하이퍼파라미터를 추가할 수 있다.
<br>이런 데이터 준비 단계를 자동화할수록 더 많은 조합을 자동으로 시도해볼 수 있고 최상의 조합을 찾을 가능성을 매우 높여준다.(그리고 시간도 많이 절약된다.)

**특성 스케일링**

머신러닝 알고리즘은 입력 숫자 특성들의 스케일이 많이 다르면 잘 작동하지 않는다.
<br>주택 가격 데이터도 이에 해당한다.
<br>즉, 전체 방 개수의 범위는 6에서 39,320인 반면 중간 소득의 범위는 0에서 15까지이다.
<br>타겟값에 대한 스케일링은 일반적으로 불필요하다.


모든 특성의 범위를 같도록 만들어주는 방법으로 min-max scaling과 표준화(standardization)가 널리 사용된다.


min-max scaling이 가장 간단하다.(많은 사람들이 이를 정규화(nomalizaton)라고 부른다.)
<br>(사이킷런에는 입력 데이터에서 각 행의 L2 norm이 1이 되도록 조정하는 Normalizer 라는 전처리 기능이 있으며 
<br>min-max scaling과는 전혀 다른 결과를 만든다. 정규화는 여러 의미로 다양하게 사용되므로 혼동하지 말자)
<br>0 ~ 1 범위에 들도록 값을 이동하고 스케일을 조정하면 된다.
<br>데이터에서 최소값을 뺀 후 최대값과 최소값의 차이로 나누면 이렇게 할 수 있다.
<br>사이킷런에는 이에 해당하는 MinMaxScaler변환기를 제공한다.
<br>0 ~ 1 사이를 원하지 않는다면 feature_range 매개변수로 범위를 변경할 수 있다.


표준화는 다르다.
<br>먼저 평균을 뺀 후(그래서 표준화를 하면 항상 평균이 0이 된다) 표준편차로 나누어 결과 분포의 분산이 1이 되도록 한다.
<br>min-max scaling과는 달리 표준화는 범위의 상한과 하한이 없어 어떤 알고리즘에서는 문제가 될 수 있다.
<br>(예를 들어 신경망은 종종 입력값의 범위로 0에서 1사이를 기대한다)
<br>그러나 표준화는 이상치에 영향을 덜 받는다. 
<br>예를 들어 중간 소득을 (잘못해서) 100이라 입력한 구역을 가정해 보자.
<br>min-max scaling은 0 ~ 15 사이의 모든 다른 값을 0 ~ 0.15로 만들어버리겠지만, 표준화는 크게 영향받지 않는다.
<br>사이킷런에는 표준화를 위한 StandardScaler 변환기가 있다.


모든 변환기에서 스케일링은 (테스트셋이 포함된) 전체 데이터가 아니고 훈련 데이터에 대해서만 fit()을 적용해야 한다.
<br>그런 다음 훈련셋과 테스트셋(그리고 새로운 데이터)에 대해 transform()를 사용한다.

**변환 파이프 라인**

앞서 보았듯이 변환 단계가 많으며 정확한 순서대로 실행되어야 한다.
<br>다행이 사이킷런에는 연속된 변환을 순서대로 처리할 수 있는 Pipeline 클래스가 있다.
<br>다음은 숫자 특승을 처리하는 간단한 파이프라인이다.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

housing_num_tr = num_pipeline.fit_transform(housing_num)
```

```python
housing_num_tr
```

Pipeline은 연속된 단계를 나타내는 이름/추정기 쌍의 목록으로 입력을 받는다.
<br>마지막 단계에는 변환기와 추정기를 모두 사용할 수 있고 그 외에는 모두 변환기여야 한다.
<br>(즉, fit_transform()을 가지고 있어야 한다. 또는 fit()과 transform() 만 가지고 있어도 된다.)
<br>이름은 (고유하고 이중 밑줄 문자인 `__`를 포함하지 않는다면) 무엇이든 상관없다.
<br>나중에 하이퍼파라미터 튜닝할 때 필요할 것이다.


파이프라인의 fit()을 호출하면 모든 변환기의 fit_transform()을 순서대로 호출하면서 
<br>(fit_transform()이 없다면 fit()과 transform()을 차례로 호출한다.) 한 단계의 출역을 다음 단계의 입력으로 전달한다.
<br>마지막 단계에서는 fit()만 호출한다.


파이프라인 객체는 마지막 추정기와 동일한 메서드를 제공한다.
<br>이 예에서는 마지막 추정기가 변환기 StandardScaler이므로 파이프라인이 데이터에 대해 모든 변환을 순서대로 적용하는 transform()를 가지고 있다.(또한 앞에서 사용한 fit_transform()도 가지고 있다.)


지금까지 범주형 열과 수치형 열을 각각 다루었다.
<br>하나의 변환기로 각 열마다 적절한 변환을 적용하여 모든 열을 처리할 수 있다면 더 편리할 것이다.
<br>사이킷런 버전 0.20에서는 이런 기능을 위해 ColumnTransformer가 추가되었다.

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", OneHotEncoder(), cat_attribs),
    ])

housing_prepared = full_pipeline.fit_transform(housing)
```

```python
housing_prepared
```

```python
housing_prepared.shape
```

ColumnTransformer의 객체는 튜플의 리스트를 생성자의 인자로 받는다.
<br>각 튜플은 이름(이중 밑줄 문자를 포함하지 않는 한 어떤 것도 가능), 변환기, 변환기가 적용될 열 이름(또는 인덱스)의
<br>리스트로 이루어진다.
<br>이 예에서 수치형 열은 앞서 정의한 num_pipeline을 사용해 변환되고 범주형 열은 OneHotEncoder을 사용해 변환되어야 한다.
<br>마지막으로 ColumnTransformer를 주택 데이터에 적용한다.
<br>각 변환기를 적절한 열에 적용하고 그 결과를 두 번째 축을 따라 연결한다.(변환기는 동일한 개수의 행을 반환해야 한다.)


OneHotEncoder는 희소행렬을 반환하지만 num_pipeline은 밀집행렬을 반환한다.
<br>희소행렬과 밀집행렬이 섞여 있을 때 ColumnTransformer는 최종 행렬의 밀집 정도를 추정한다.(즉, 0이 아닌 원소의 비율)
<br>밀집도가 임계값(기본적으로 sparse_threshold=0.3) 보다 낮으면 희소행렬을 반환한다.
<br>이 예에서는 밀집행렬이 반환된다.


튜플에 변환기를 사용하는 대신 삭제하고 싶은 열이 있다면 "drop" 문자열로 지정하거나 
<br>변환을 적용하지 않을 열이 있다면 "passthrough" 로 지정할 수 있다. 
<br>기본적으로 나머지 열(즉, 나열되지 않은 열)은 삭제된다.
<br>하지만 이런 열을 다르게 처리하고 싶다면 remainder 하이퍼파라미터에 어떤 변환기를 (또는 "passthrough"를) 지정할 수 있다.
<br>(reaminder 파라미터의 기본값은 "drop"이다.)

```python
# 다음은 (판다스 DataFrame 열의 일부를 선택하기 위해) DataFrameSelector 변환기와 FeatureUnion를 사용한 예전 방식이다:
# 각 변환기에 열을 따로 지정할 수 없고 전체 데이터에 모두 적용된다.
# 이 문제를 해결하기 위해 열을 선택해주는 사용자 정의 변환기를 아래와 같이 만들 수 있다.
from sklearn.base import BaseEstimator, TransformerMixin

# 수치형 열과 범주형 열을 선택하기 위한 클래스
class OldDataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
# 하나의 큰 파이프라인에 이들을 모두 결합하여 수치형과 범주형 특성을 전처리한다:
num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

old_num_pipeline = Pipeline([
        ('selector', OldDataFrameSelector(num_attribs)),
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

old_cat_pipeline = Pipeline([
        ('selector', OldDataFrameSelector(cat_attribs)),
        ('cat_encoder', OneHotEncoder(sparse=False)),
    ])
```

```python
from sklearn.pipeline import FeatureUnion

old_full_pipeline = FeatureUnion(transformer_list=[
        ("num_pipeline", old_num_pipeline),
        ("cat_pipeline", old_cat_pipeline),
    ])
```

```python
# ColumnTransformer의 결과와 동일하다:
old_housing_prepared = old_full_pipeline.fit_transform(housing)
old_housing_prepared
```

```python
np.allclose(housing_prepared, old_housing_prepared)
```

**교차 검증을 사용한 평가**

우선은 수작업으로 train_test_split()를 사용해 훈련세트를 더 작은 훈련세트와 검증 세트로 나누고, 
<br>더 작은 훈련세트에서 모델을 훈련시키고 검증 세트로 모델을 평가하는 방법이 있다.
<br>두 번째는 사이킷런의 k fold cross validation 기능을 사용하는 방법도 있다.

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                         scoring="neg_mean_squared_error", cv=10)
tree_rmse_scores = np.sqrt(-scores)
```

```python
# 결과를 살펴보자.
def display_scores(scores):
    print("점수:", scores)
    print("평균:", scores.mean())
    print("표준 편차:", scores.std())

display_scores(tree_rmse_scores)
```

사이킷런의 교차 검증 기능은 scoring 매개변수에 (낮을수록 좋은) 비용함수가 아니라 (클수록 좋은) 효용함수를 기대한다.
<br>그래서 평균제곱오차(MSE)의 반대값(즉, 음수값)을 계산하는 neg_mean_squared_error를 사용한다.
<br>이런 이유로 앞선 코드에서 제곱근을 계산하기 전에 -scoring으로 부호를 바꿨다.
<br>(평균제곱오차가 작을수록 좋은 비용함수이므로 부호가 반대가 되어야 scoring 매개변수 정의에 맞는다.
<br>회귀모델에서 scoring 매개변수를 지정하지 않으면 기본적으로 0 ~ 1 사이의 값을 가지는 r2_score가 사용된다.)


결과는 이전만큼 좋아보이지 않는다. 실제로 선형회귀모델보다 나쁘다.
<br>교차검증으로 모델의 성능을 추정하는 것뿐만 아니라 이 추정이 얼마나 정확한지(즉, 표준편차) 측정할 수 있다.
<br>결정 트리 점수가 대략 평균 71,407 에서 ±2,439 사이이다.
<br>검증 세트를 하나만 사용했다면 이런 정보를 얻지 못했을 것이다.
<br>하지만 모델을 여러 번 훈련시켜야 해서 비용이 비싸므로 교차검증을 언제나 쓸 수 있는 것은 아니다.

**그리드 탐색**

가장 단순한 방법은 만족할 만한 하이퍼파라미터 조합을 찾을 때까지 수동으로 하이퍼파라미터를 조정하는 것이다.
<br>이는 매우 지루한 작업이고 많은 경우의 수를 탐색하기에는 시간이 부족할 수도 있다.
<br>
<br>
대신 사이킷런의 GridSearchCV를 사용하는 것이 좋다.

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    # 12(=3×4)개의 하이퍼파라미터 조합을 시도한다.
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # bootstrap은 False로 하고 6(=2×3)개의 조합을 시도한다.
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestRegressor(random_state=42)

# 다섯 개의 폴드로 훈련하면 총 (12+6)*5=90번의 훈련이 일어난다.
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)
```

어떤 하이퍼파라미터 값을 지정해야 할지 모를 때는 연속된 10의 거듭제곱 수로 시도해보는 것도 좋다.
<br>(더 세밀하게 탐색하려면 위 예제의 n_estimators 하이퍼파라미터처럼 더 작은 값을 지정한다.)

```python
# 최상의 파라미터 조합은 다음과 같다.
# 8과 30은 탐색 범위의 최대값이기 때문에 계속 점수가 향상될 수 있으므로 더 큰 값으로 다시 검색해야 한다.
grid_search.best_params_
```

```python
# 최적의 추정기에 직접 접근할 수도 있다.
grid_search.best_estimator_
```

GridSearchCV가 (기본값인) refit=True로 초기화 되었다면 교차 검증으로 최적의 추정기를 찾은 다음 전체 훈련 세트로 다시 훈련시킨다.
<br>일반적으로 데이터가 많을 수록 성능이 향상되므로 좋은 방법이다.

```python
# 평가 점수 확인
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

<br>하지만 하이퍼파라미터 탐색 공간이 커지면 RandomizedSearchCV를 사용하는 편이 더 좋다.
<br>(특히 규제처럼 설정값이 연속형인 경우 랜덤 탐색이 권장된다.)

<br>주요 장점은 다음과 같다.
* 랜덤 탐색을 1,000회 반복하도록 실행하면 하이퍼파라미터마다 각기 다른 1,000개의 값을 탐색한다.
<br>(그리드 탐색에서는 하이퍼파라미터마다 몇 개의 값만 탐색한다.)
* 단순히 반복 횟수를 조정하는 것만으로 하이퍼파라미터 탐색에 투입할 컴퓨팅 자원을 제어할 수 있다.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=8),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(housing_prepared, housing_labels)
```

```python
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
```

랜덤 탐색을 10회 반복하여 찾은 최적의 하이퍼파라미터는 n_estimators가 180, max_features가 7이다.

<br>현재 제품 시스템에 있는 모델보다 불과 0.1% 더 높다면 어떨까? 이 추정값이 얼마나 정확한지 알고 싶을 것이다.
<br>이를 위해 scipy.stats.t.interval()를 사용해 일반화 오차의 95% 신뢰구간(confidence interval)을 계산할 수 있다.

```python
from scipy import stats

confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))
```

```python
# 또는 다음과 같이 수동으로 계산할 수도 있다.
m = len(squared_errors)
mean = squared_errors.mean()
tscore = stats.t.ppf((1 + confidence) / 2, df=m - 1)
tmargin = tscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - tmargin), np.sqrt(mean + tmargin)
```

```python
# 또는 t-점수 대신 z-점수를 사용할 수도 있다.
zscore = stats.norm.ppf((1 + confidence) / 2)
zmargin = zscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - zmargin), np.sqrt(mean + zmargin)
```

하이퍼파라미터 튜닝을 많이 했다면 교차 검증을 사용해 측정한 것보다 조금 성능이 낮은 것이 보통이다.
<br>(우리 시스템이 검증 데이터에서 좋은 성능을 내도록 세밀하게 튜닝되었기 때문에 새로운 데이터셋에는 잘 작동하지 않을 가능성이 크다.)
<br>이 예제에서는 성능이 낮아지지 않았지만, 이런 경우가 생기더라도 테스트셋에서 성능 수치를 좋게 하려고 하이퍼파라미터를 튜닝하려 시도해서는 안된다.
<br>그렇게 향상된 성능은 새로운 데이터에 일반화되기 어렵다.


이제 프로젝트 런칭 직전 단계에 왔다.
<br>(학습한 것, 한 일과 하지 않은 일, 수립한 가정, 시스템 제한 사항 등을 강조하면서) 솔루션과 문서를 출시하고, 
<br>깔끔한 도표와 기억하기 쉬운 제목으로 (예를 들면 '수입의 중간값이 주택 가격 예측의 가장 중요한 지표다') 멋진 발표 자료를 만들어야 한다.
<br>캘리포니아 주택 가격 예제에서 시스템의 최종 성능이 전문가의 가격 예측보다 좋지 않다. 종종 20%까지 차이가 난다.
<br>하지만 전문가에게 여유를 주어 더욱 흥미롭고 생산적인 작업을 할 수 있다면 이 시스템을 런칭하는 것이 좋다.

**1. 서포트 벡터 머신 회귀(sklearn.svm.SVR)를 kernel=“linear”(하이퍼파라미터 C를 바꿔가며)나 
<br>kernel=“rbf”(하이퍼파라미터 C와 gamma를 바꿔가며) 등의 다양한 하이퍼파라미터 설정으로 시도해보세요. 
<br>지금은 이 하이퍼파라미터가 무엇을 의미하는지 너무 신경 쓰지 마세요. 
<br>최상의 SVR 모델은 무엇인가요?**

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
        {'kernel': ['linear'], 'C': [10., 30., 100., 300., 1000., 3000., 10000., 30000.0]},
        {'kernel': ['rbf'], 'C': [1.0, 3.0, 10., 30., 100., 300., 1000.0],
         'gamma': [0.01, 0.03, 0.1, 0.3, 1.0, 3.0]},
    ]

svm_reg = SVR()
grid_search = GridSearchCV(svm_reg, param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2)
grid_search.fit(housing_prepared, housing_labels)
```

```python
# 최상 모델의 (5-폴드 교차 검증으로 평가한) 점수는 다음과 같다:
negative_mse = grid_search.best_score_
rmse = np.sqrt(-negative_mse)
rmse
```

```python
# RandomForestRegressor보다 훨씬 좋지 않ek. 최상의 하이퍼파라미터를 확인해 보겠다:
grid_search.best_params_
```

선형 커널이 RBF 커널보다 성능이 나은 것 같다. C는 테스트한 것 중에 최대값이 선택되었다. 
<br>따라서 (작은 값들은 지우고) 더 큰 값의 C로 그리드서치를 다시 실행해 보아야 한다. 
<br>아마도 더 큰 값의 C에서 성능이 높아질 것이다.

**2. GridSearchCV를 RandomizedSearchCV로 바꿔보세요.**

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import expon, reciprocal

# expon(), reciprocal()와 그외 다른 확률 분포 함수에 대해서는
# https://docs.scipy.org/doc/scipy/reference/stats.html를 참고하세요.

# 노트: kernel 매개변수가 "linear"일 때는 gamma가 무시됩니다.
param_distribs = {
        'kernel': ['linear', 'rbf'],
        'C': reciprocal(20, 200000),
        'gamma': expon(scale=1.0),
    }

svm_reg = SVR()
rnd_search = RandomizedSearchCV(svm_reg, param_distributions=param_distribs,
                                n_iter=50, cv=5, scoring='neg_mean_squared_error',
                                verbose=2, random_state=42)
rnd_search.fit(housing_prepared, housing_labels)
```

```python
# 최상 모델의 (5-폴드 교차 검증으로 평가한) 점수는 다음과 같다:
negative_mse = rnd_search.best_score_
rmse = np.sqrt(-negative_mse)
rmse
```

```python
# 이제 RandomForestRegressor의 성능에 훨씬 가까워졌다(하지만 아직 차이가 난다). 
# 최상의 하이퍼파라미터를 확인해 보겠다:
rnd_search.best_params_
```

이번에는 RBF 커널에 대해 최적의 하이퍼파라미터 조합을 찾았다. 
<br>보통 랜덤서치가 같은 시간안에 그리드서치보다 더 좋은 하이퍼파라미터를 찾는다.


여기서 사용된 scale=1.0인 지수 분포를 살펴보겠다. 일부 샘플은 1.0보다 아주 크거나 작다. 
<br>하지만 로그 분포를 보면 대부분의 값이 exp(-2)와 exp(+2), 즉 0.1과 7.4 사이에 집중되어 있음을 알 수 있다.

```python
expon_distrib = expon(scale=1.)
samples = expon_distrib.rvs(10000, random_state=42)
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Exponential distribution (scale=1.0)")
plt.hist(samples, bins=50)
plt.subplot(122)
plt.title("Log of this distribution")
plt.hist(np.log(samples), bins=50)
plt.show()
```

C에 사용된 분포는 매우 다르다. 주어진 범위안에서 균등 분포로 샘플링된다. 
<br>그래서 오른쪽 로그 분포가 거의 일정하게 나타난다. 
<br>이런 분포는 원하는 스케일이 정확이 무엇인지 모를 때 사용하면 좋다:

```python
reciprocal_distrib = reciprocal(20, 200000)
samples = reciprocal_distrib.rvs(10000, random_state=42)
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.title("Reciprocal distribution (scale=1.0)")
plt.hist(samples, bins=50)
plt.subplot(122)
plt.title("Log of this distribution")
plt.hist(np.log(samples), bins=50)
plt.show()
```

reciprocal() 함수는 하이퍼파라미터의 스케일에 대해 전혀 감을 잡을 수 없을 때 사용한다
<br>(오른쪽 그래프에서 볼 수 있듯이 주어진 범위안에서 모든 값이 균등하다). 
<br>반면 지수 분포는 하이퍼파라미터의 스케일을 (어느정도) 알고 있을 때 사용하는 것이 좋다.
